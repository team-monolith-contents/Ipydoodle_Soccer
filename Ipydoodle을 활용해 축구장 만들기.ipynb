#### 2.1.2. 좌표 판정하기

* 공의 중심이 경기장 안에 있으면 "공이 안에 있습니다"를 출력합니다.
* 공의 중심이 경기장 경계선 위에 있으면 안에 있다고 판정합니다.
* 조건문은 다음과 같이 사용합니다.
    ```python
    if ("조건문"):
        # 조건문이 참일때 실행할 코드
    else:
        # 조건문이 거짓을때 실행할 코드
    ```
* 비교 연산자는 다음과 같이 사용합니다.
    ```python
    print(1 < 2) # 1이 2보다 작으므로 True를 출력합니다.
    print(2 < 1) # 1이 2보다 작지 않으므로 False를 출력합니다.
    ```
* 논리연산자는 다음과 같이 사용합니다. 논리연산자의 규칙은 다음과 같습니다
|   a   |   b   | a and b | a or b | not a |
|:-----:|:-----:|:-------:|--------|-------|
|  True |  True |   True  | True   | False |
|  True | False |  False  | True   | False |
| False | True  | False   | True   | True  |
| False | False | False   | False  | True  |
* 예시로 다음과 같이 사용합니다
    ```python
    print(1 < 2 and 3 < 4) # 관계연산이 모두 참이므로, True를 출력합니다.
    print(1 < 2 and 3 > 4) # 두번째 관계연산이 거짓이므로, False를 출력합니다.
    print(1 < 2 or 3 > 4) # 첫번째 관계연산이 참이므로, True를 출력합니다.
    ```

##### 2.1.2. 해설
공이 바깥에 있을 조건은 중심의 좌표가 경계선 바깥에 있는 경우입니다.  
중심 좌표가 경계선 바깥인 경우는 크게 4가지 경우가 있습니다.
1. x 좌표가 300보다 큰 경우
2. x 좌표가 -300보다 작은 경우
3. y 좌표가 150보다 큰 경우
4. y 좌표가 -150보다 작은 경우

## 3. IpyDoodle과 함께 반복 구조 연습하기

### 3.1. 공이 여러개라면?

철수는 작성한 코드가 여러 상황에 대해 완벽히 작동하는지 공을 여러개 만들어 경기장에 생성시키려고 합니다.  
축구공의 중심 좌표가 여러개 주어졌을 때, 여러분이 짠 코드는 완벽히 작동할까요?

* x 좌표이 담긴 1차원 배열 xs와 y좌표가 담긴 1차원 배열 ys가 주어집니다.
* x좌표와 y좌표가 함께 주어질 때, 각 배열에 대해 공이 안에 있는지 혹은 밖에 있는지를 판정합니다.
    * 공이 안에 있는 경우에만 공을 그리고 나머지 경우는 그리지 않습니다.
    * 공이 바깥에 있는 경우, "공이 바깥에 있습니다"를 출력합니다.
* 배열은 다음과 같이 사용합니다.
    ```python
    xs [1, 2, 3, 4, 5, 6]
    print(xs[0]) # 배열의 1번째 원소를 출력합니다. 지금의 경우에는 1을 출력합니다.
    print(xs[3]) # 배열의 4번째 원소를 출력합니다. 지금의 경우에는 4을 출력합니다.
    ```
* 반복문은 다음과 같이 사용합니다
    ```python
    xs = [1, 2, 3, 4, 5, 6]
    for i in range(6): # 배열의 길이가 들어갑니다.
        print(xs[i]) # 1~6까지 출력됩니다
    ```

````python
"""주어진 코드
xs = [0, 300, -301, 300, -299]
ys = [0, 150, 150, -150, -151]
"""

xs = [0, 300, -301, 300, -299]
ys = [0, 150, 150, -150, -151]

World(width = 700, height = 400)
Box(x = 0, y = 0, width = 600, height = 300, color='green')

for i in range(5):
    x = xs[i]
    y = ys[i]
    if x > 300 or x < -300 or y > 150 or y < -150:
        print("공이 바깥에 있습니다.")
    else:
        Circle(x = x, y = y, radius = 25, color='black')
````

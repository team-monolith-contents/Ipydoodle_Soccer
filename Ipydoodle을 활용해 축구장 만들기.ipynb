## 3. IpyDoodle    

### 3.1.  ?

             .  
     ,     ?

* x   1  xs y  1  ys .
* x y   ,          .
    *          .
    *    , "  " .
*    .
    ```python
    xs [1, 2, 3, 4, 5, 6]
    print(xs[0]) #  1  .   1 .
    print(xs[3]) #  4  .   4 .
    ```
*    
    ```python
    xs = [1, 2, 3, 4, 5, 6]
    for i in range(6): #   .
        print(xs[i]) # 1~6 
    ```

````python
""" 
xs = [0, 300, -301, 300, -299]
ys = [0, 150, 150, -150, -151]
"""

xs = [0, 300, -301, 300, -299]
ys = [0, 150, 150, -150, -151]

World(width = 700, height = 400)
Box(x = 0, y = 0, width = 600, height = 300, color='green')

for i in range(5):
    x = xs[i]
    y = ys[i]
    if x > 300 or x < -300 or y > 150 or y < -150:
        print("  .")
    else:
        Circle(x = x, y = y, radius = 25, color='black')
````
